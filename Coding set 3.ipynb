```python
 #3.Find Even and Odd Numbers in a List
# Question: Write a Python program to separate even and odd numbers from the list [12, 17, 19, 24, 35, 40, 56].
def Find_num(list_1):
    odd=[]
    even=[]
    for num in list_1:
        if num%2==0:
            even.append(num)
            
            
        else:
            odd.append (num)
    print(f"Odd numbers in the given list : \n{odd}\n")
    print(f"Even numbers in the given list : \n{even}")
list_1=[12, 17, 19, 24, 35, 40, 56]
Find_num(list_1)

```
